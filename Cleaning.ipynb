```
url2='https://raw.githubusercontent.com/RaghavSinghal10/DS203_Project/main/covid_19_india.csv'
response = requests.get(url2)

raw_data = response.content.decode('utf-8')
raw_data = raw_data.split("\n")


with open('covid_india.csv', 'w') as file:
  for row in raw_data:
    file.write(row+"\n")
df_cov_ind = pd.read_csv("covid_india.csv")
```

```
url3='https://raw.githubusercontent.com/RaghavSinghal10/DS203_Project/main/time_series_covid_19_confirmed.csv'
response = requests.get(url3)

raw_data = response.content.decode('utf-8')
raw_data = raw_data.split("\n")


with open('confirmed.csv', 'w') as file:
  for row in raw_data:
    file.write(row+"\n")
df_confirmed = pd.read_csv("confirmed.csv")

df_confirmed = df_confirmed.drop(columns = ["Province/State"])
df_confirmed = df_confirmed.rename(columns={"Country/Region": "Country"})
```

```
df_confirmed.isnull().sum().sum()
```

```
df_confirmed = df_confirmed.groupby("Country").sum()
```

```
df_confirmed = df_confirmed.drop(columns = ["Lat","Long"]) 
```

```
df_confirmed_1 = df_confirmed.transpose()
```

```
df_confirmed_1
```

```
df_confirmed_1 = df_confirmed_1.reset_index()
df_confirmed_1 = df_confirmed_1.rename(columns = {"index" : "date"})
df_confirmed_1["date"] = pd.to_datetime(df_confirmed_1["date"])
df_confirmed_1
```

```
url4='https://raw.githubusercontent.com/RaghavSinghal10/DS203_Project/main/time_series_covid_19_deaths.csv'
response = requests.get(url4)

raw_data = response.content.decode('utf-8')
raw_data = raw_data.split("\n")


with open('deaths.csv', 'w') as file:
  for row in raw_data:
    file.write(row+"\n")
df_deaths = pd.read_csv("deaths.csv")
```

```
df_deaths = df_deaths.drop(columns = ["Province/State","Lat","Long"])
df_deaths = df_deaths.rename(columns={"Country/Region": "Country"})
df_deaths = df_deaths.groupby("Country").sum()
df_deaths = df_deaths.transpose()
df_deaths = df_deaths.reset_index()
df_deaths = df_deaths.rename(columns = {"index" : "date"})
df_deaths["date"] = pd.to_datetime(df_deaths["date"])
```

```
df_deaths
```

```
url5='https://raw.githubusercontent.com/RaghavSinghal10/DS203_Project/main/time_series_covid_19_recovered.csv'
response = requests.get(url5)

raw_data = response.content.decode('utf-8')
raw_data = raw_data.split("\n")


with open('recovered.csv', 'w') as file:
  for row in raw_data:
    file.write(row+"\n")
df_recovered = pd.read_csv("recovered.csv")

df_recovered = df_recovered.drop(columns = ["Province/State","Lat","Long"])
df_recovered = df_recovered.rename(columns={"Country/Region": "Country"})
df_recovered = df_recovered.groupby("Country").sum()
df_recovered = df_recovered.transpose()
df_recovered = df_recovered.reset_index()
df_recovered = df_recovered.rename(columns = {"index" : "date"})
df_recovered["date"] = pd.to_datetime(df_recovered["date"])
df_recovered
```

```
df_deaths.to_csv('time_series_covid_deaths.csv') 
files.download('time_series_covid_deaths.csv')
df_confirmed_1.to_csv('time_series_covid_confirmed.csv') 
files.download('time_series_covid_confirmed.csv')
df_recovered.to_csv('time_series_covid_recovered.csv') 
files.download('time_series_covid_recovered.csv')
df_cov_ind.to_csv('covid_india.csv') 
files.download('covid_india.csv')
```
